Distance in megaparsecs (Mpc) on the x-axis, and Velocity in kilometers per second (km/s) on the y-axis

```python
import numpy as np
import matplotlib.pyplot as plt #always start with numpy
```

```python
X = np.array([19,65,66,80,97,113,175,270,310,350,350,650,680])    #importing data x being the distnace in mpc
Y = np.array([1150,3800,5000,4800,5400,6700,10300,15400,19500,23200,21600,39400,41000])  #and y being the speed in km/s
```

```python
plt.plot(X,Y,'g.') #plotting out data
```

```python
m_guess = 0
b_guess = 0 #guesses a slope and y intercept value
```

$y=mx+b$

```python
m_guess * X + b_guess #guesses that the value will be 0 of course because if the slope is 0 and the intercept is 0, the x value will always be 0
```

```python
y_guess = m_guess * X + b_guess #slope formula
```

```python
y_guess - Y #guesses a y value and subtracts it by the np.array provided earlier 
```

```python
np.sum((y_guess - Y)**2) #finds the sum of the equation
```

```python
loss = np.sum((y_guess - Y)**2) #loss is the predicted output-actual output
```

```python
best_loss_so_far = 1000000000000
m_best = 0 
b_best = 0 
for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):
        
        y_guess = m_guess * X + b_guess

        loss = np.sum((y_guess - Y)**2)

        if loss < best_loss_so_far:
            m_best = m_guess
            b_best = b_guess
            best_loss_so_far = loss

print("Found best fit:")
print(m_best,b_best,loss) #prints the best fit 
```

```python
m_best #shows best slope fit 
```

```python
H = m_best #the variable for the best fit slope is called H
```

```python
1/H #1/61.6
```
