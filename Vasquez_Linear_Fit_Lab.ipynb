```python
import numpy as np
import matplotlib.pyplot as plt #the blueprint
```

```python
x = np.linspace(-10,10,20) #creates 20 numbers that range form -10 to 10
```

```python
x #lists the numbers
```

```python
m = 50 #makes the slope 50
b = 17 #makes the y intercept 17
```

```python
y = m * x + b #formula for y intercept
```

```python
plt.plot(x,y,'.') #plots the formula onto a dotted line graph
```

```python
r = 100*np.random.randn(20,) #generates 20 random numbers with no range
```

```python
r #lists the 20 random numbers 
```

```python
plt.plot(r,'.') #plots the 20 random number on a dotted graph
```

```python
y2 = m * x + b + r #doubles the y values and adds r to the y intercept
```

```python
plt.plot(x,y2,'.') #puts the data on a dotted graph
```

```python
m_guess = 1 #slope
b_guess = 1 #y intercept

y_guess = m_guess * x + b_guess #slope formula

plt.plot(x,y_guess,'r.'); #line of best fit 
plt.plot(x,y2,'g.'); #plots the data onto a green dotted graph
```

```python
for m_guess in np.linspace(-100,100,100):#guesses a 100 different numbers for the slope that ranges from -100 to 100
    
        y_guess = m_guess * x 

        plt.plot(x,y_guess,'r.'); 
        plt.plot(x,y2,'g.'); #plots the data
        plt.show() #shows the data

        loss = np.sum((y_guess - y)**2) #shows much the predicted amount differs from the actual

        print(loss,'M = ',m_guess) #prints the results
```

```python
best_loss = 100000000000
best_m = 0
best_b = 0 

for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):
        
        y_guess = m_guess * x + b_guess #guesses a random slope formula 

        loss = np.sum((y_guess - y)**2) #tells us the predicted output of the model differs from the actual output

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess #prints best slope value
            best_b = b_guess #prints best y intercept value
```

```python
y_guess = best_m * x + best_b #slope formula

plt.plot(x,y_guess,'r.'); #plots using red dots
plt.plot(x,y2,'g.'); #plots using green dots
plt.show() #plots data on dotted graph

print(best_loss,best_m,best_b) #prints the best predicted results
```
